```r
COL0 = df$Global_Sales[df$Genre == 'Action']
COL1 = df$Global_Sales[df$Genre == 'Sports']

varianceCOL0 <- var(COL0)
varianceCOL1 <- var(COL1)
difference <- diff(varianceCOL0, varianceCOL1)
varianceCOL1
varianceCOL0

#La valeur d'une variance ne peut être interprétée que par comparaison à la valeur d'une norme ou d'une autre variance. Si une variance est nulle, cela veut dire que toutes les observations sont égales à la moyenne, ce qui implique qu'il n'y a aucune variation de celles-ci.
```

```r
COl0_moyenne <- mean(COL0)
COl1_moyenne <- mean(COL1)
COl0_moyenne
COl1_moyenne
# On remarque la moyenne est a peu près pareille pour les 2, ce qui veut dire que parmis les jeux proposés, la mojorité sont des jeux de sports.
```

```r
size_sum(COL0)
size_sum(COL1)
dim(COL0)
summary(COL0)
summary(COL1)

# On observe que les mins sont les mêmes; mais que les Medians, Moyenne, 1st Quartile et 3rd Quartile et Max value sont different.
```

```r
wilcox.test(COL0, conf.int = TRUE)
wilcox.test(COL1, conf.int = TRUE)
```

```r
COL_adventure = df$Global_Sales[df$Genre == 'Adventure']
COL_strategy  = df$Global_Sales[df$Genre == 'Strategy']
shapiro.test(COL_adventure)
shapiro.test(COL_strategy)
COL_adventure
# p-value < 2.2e-16 donc on rejette l'hypothèse H0. Cela veut dire que les 2 ne sont pas distribué suivant une loi normale
```

```r
myfun <- function(x, group) {
  data.frame(x, group) %>%
  group_by(group) %>%
  summarise(
    statistic = ifelse(sd(x)!=0,shapiro.test(x)$statistic,NA), 
    p.value = ifelse(sd(x)!=0,shapiro.test(x)$p.value,NA)
  )
}
(lst <- lapply(df[,-1], myfun, group=df[,1]))

```
